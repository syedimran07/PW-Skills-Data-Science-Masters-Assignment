### Q1. Explain GET and POST methods.

Ans: The GET method in HTTP or Hypertext Transfer Protocol, is designed for the purpose of retrieving data from a specified resource. When using GET, the data is conveyed in the URL as a query string, which is appended after a question mark. This method is commonly employed for straightforward data retrieval tasks, where the parameters are visible in the URL. The visibility of data in the URL makes it suitable for simple requests, but it's important to note that this method is best suited for idempotent and safe operations. Additionally, GET requests have the capability to be cached by browsers, contributing to their effectiveness in scenarios where data retrieval does not modify the server state. 


The POST method in HTTP (Hypertext Transfer Protocol) serves the purpose of submitting data to be processed by a specified resource. In contrast to GET, where data is appended to the URL as a query string, with POST, the data is transmitted within the body of the HTTP request. This method is particularly useful for handling sensitive or large amounts of data, as the parameters are not visible in the URL, enhancing privacy and security. POST requests are commonly employed for non-idempotent and potentially unsafe operations, where the intention is to modify the server state. Unlike GET, POST requests are typically not cached by browsers, emphasizing their suitability for situations where data submission and processing are involved. 

### Q2. Why is request used in Flask?

Ans: In Flask, the "request" object is used to handle incoming HTTP requests made by clients (typically web browsers) to the server. Flask is a micro web framework for Python, and the "request" object provides a convenient way to access and process information sent by the client in the HTTP request.

### Q3. Why is redirect() used in Flask?

Ans: 
In Flask, the redirect() function is employed to smoothly redirect users to a different URL, enhancing the navigation flow within a web application. This function simplifies URL construction and maintenance, promoting cleaner and more readable code. Additionally, it aids in preventing form resubmission issues and supports dynamic redirection based on application conditions.

### Q4. What are templates in Flask? Why is the render_template() function used?

Ans: In Flask, "templates" refer to files containing HTML code with embedded placeholders for dynamic content. These placeholders are replaced with actual data during runtime, allowing developers to create dynamic web pages by combining HTML structure with Python logic. Flask uses the Jinja2 templating engine to achieve this integration of server-side logic within HTML templates. Templates play a crucial role in separating the presentation layer from the application logic, promoting code organization and maintainability. By utilizing templates, developers can easily generate dynamic content, handle conditionals and loops, and create reusable components, providing a clean and efficient way to build web interfaces in Flask applications.

The render_template() function in Flask is used to combine HTML templates with dynamic content, generated by processing placeholders with the Jinja2 templating engine. It enables the creation of dynamic web pages by injecting server-side data into the HTML structure. This function supports code organization, separation of concerns, and the development of consistent and reusable components for web applications.

### Q5. Create a simple API. Use Postman to test it. Attach the screenshot of the output in the Jupyter Notebook.

```python
from flask import Flask, request, Response, json, jsonify

app = Flask(__name__)

@app.route('/get-result', methods = ['POST'])
def get_even_or_odd():
    number = request.get_json()
    num = int(number['num'])
    if num % 2 == 0:
        result = "Even Number"
    else:
        result = "Odd Number"
    return jsonify(result)

if __name__ == "__main__":
    app.run(debug=True)
```
